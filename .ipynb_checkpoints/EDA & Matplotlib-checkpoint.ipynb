```python
#imoorting libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
import plotly.graph_objects as go
import plotly.express as px
import datetime
from plotly.offline import init_notebook_mode, iplot
from plotly.subplots import make_subplots
init_notebook_mode(connected=True)
```

```python
#loading data
data= pd.read_csv("C:/Users/works/Desktop/Tech stack/projects-20230901T120436Z-001/projects/BA/Google-Playstore.csv")
```

## Overall app ratings

```python
# Histogram of app ratings
plt.hist(data['Rating'], bins=30, edgecolor='black')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of App Ratings')
plt.show()
```

----------

## Count of app categories

```python
# Countplot of app categories
plt.figure(figsize=(12, 6))
data['Category'].value_counts().plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Count of App Categories')
plt.xticks(rotation=90)
plt.show()
```

## App instals by category

```python
# Remove '+' and ',' symbols from the Installs column to convert it to a numerical format
data['Installs'] = data['Installs'].apply(lambda x: str(x).replace('+', '').replace(',', '') if isinstance(x, str) else str(x))
data['Installs'] = pd.to_numeric(data['Installs'], errors='coerce')

# Bar plot of app installs by category
plt.figure(figsize=(12, 6))
install_by_category = data.groupby('Category')['Installs'].sum().sort_values(ascending=False)
install_by_category.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Number of Installs')
plt.title('Number of Installs by App Category')
plt.xticks(rotation=90)
plt.show()
```

## App rating by Category

```python
# Boxplot of app ratings by category
plt.figure(figsize=(12, 6))
data.boxplot(column='Rating', by='Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.title('App Ratings by Category')
plt.xticks(rotation=90)
plt.show()
```

## App prices w.r.t category

```python
import seaborn as sns

# Convert Price column to string format and remove dollar sign
data['Price'] = data['Price'].astype(str).str.replace('$', '')

# Convert Price column to numeric format
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')

# Strip plot of app prices by category
plt.figure(figsize=(12, 6))
sns.stripplot(data=data, x='Category', y='Price', jitter=True, linewidth=0.5)
plt.xlabel('Category')
plt.ylabel('Price')
plt.title('App Prices by Category')
plt.xticks(rotation=90)
plt.show()
```

## Content ratings review

```python
# Countplot of content ratings
plt.figure(figsize=(10, 6))
data['Content Rating'].value_counts().plot(kind='bar')
plt.xlabel('Content Rating')
plt.ylabel('Count')
plt.title('Count of Apps by Content Rating')
plt.xticks(rotation=45)
plt.show()
```

## Most installed apps

```python
# Sort the data by Installs column in descending order
sorted_data = data.sort_values(by='Installs', ascending=False)

# Select the top 20 most installed apps
top_20_apps = sorted_data.head(20)

# Bar plot of the top 10 most installed apps
plt.figure(figsize=(12, 6))
plt.bar(top_20_apps['App Name'], top_20_apps['Installs'])
plt.xlabel('App Name')
plt.ylabel('Number of Installs')
plt.title('Top 20 Most Installed Apps')
plt.xticks(rotation=90)
plt.show()
```

```python
# Box plot of app ratings by category
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Category', y='Rating')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.title('App Ratings by Category')
plt.xticks(rotation=90)
plt.show()
```

## Top 20 apps

```python
### Sort the data by Rating column in descending order
sorted_data = data.sort_values(by='Rating', ascending=False)

# Select the top 10 highest rated apps
top_rated_apps = sorted_data.head(30)

# Bar plot of the top-rated apps
plt.figure(figsize=(12, 6))
plt.bar(top_rated_apps['App Name'], top_rated_apps['Rating'])
plt.xlabel('App Name')
plt.ylabel('Rating')
plt.title('Top-rated Apps')
plt.xticks(rotation=90)
plt.ylim(4, 5)  # Set the y-axis limit to focus on the highest ratings
plt.show()
```
